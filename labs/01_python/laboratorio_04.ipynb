EL conjunto de datos se denomina `Automobile_data.csv`, el cual contiene información tal como: compañia, precio, kilometraje, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import os
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","Automobile_data.csv")).set_index('index')
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Elimine los valores nulos (Nan)

```python
print('Vamos a reemplazar los valores de Nan por string vacio mediante la funcion .replace(), por tanto la respuesta es:')
df.replace({'Nan': ''})
```

2. Encuentra el nombre de la compañía de automóviles más cara

```python
print('Primero agruparemos las compañias mediante su precio promedio, esto es:')
precio_prom=df.groupby("company")["price"].mean()
precio_prom
```

```python
print('Luego para encontrar la compañia con precios mas cara, usamos la funcion .idxmax. Por tanto, la respuesta es:')
precio_prom=df.groupby("company")["price"].mean().idxmax()
precio_prom
```

3. Imprimir todos los detalles de Toyota Cars

```python
df.loc[df['company'] == 'toyota']
```

4. Cuente el total de automóviles por compañía

```python
#Por conveniencia dejamos el df con un indice
df = pd.read_csv(os.path.join("data","Automobile_data.csv"))
df
```

```python
print('Por tanto, la respuesta es:\n')
df.groupby('company')['index'].count()
```

5. Encuentra el coche con el precio más alto por compañía

```python
lista=[]
df_auto_mas_caro=df.groupby("company")['price'].idxmax() #entrega los autos mas caros por compañia

for marca in df.company.unique(): #se recorre el array de compañias
    lista.append(df_auto_mas_caro[marca])

print(df.iloc[lista])
```

6. Encuentre el kilometraje promedio (**average-mileage**) de cada compañía automotriz

```python
df.groupby('company')['average-mileage'].mean()
```

7. Ordenar todos los autos por columna de precio (**price**)

```python
#Acá haremos uso de la funcion .sort_values cpm los argumentos de 'price' y que ordene todos los precios de mayor a menor
df.sort_values('price',ascending=False) 
```
