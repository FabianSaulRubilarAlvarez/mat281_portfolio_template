1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
#Gráfico lineal 

palette = sns.color_palette("hls", 2)

sns.lineplot(
    x='month_number',
    y='total_profit',
    data=df,
    ci = None,
    palette=palette
)   
```

```python
#Gráfico de dispersión

#df.plot(kind='scatter',x='month_number',y='total_profit',color='green')
#plt.show()

sns.scatterplot(
    x='month_number',
    y='total_profit',
    data=df,
    palette=palette
)   
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
ax = plt.gca()

df.plot(kind='line',x='month_number',y='facecream',ax=ax)
df.plot(kind='line',x='month_number',y='facewash', color='red', ax=ax)
df.plot(kind='line',x='month_number',y='toothpaste', color='black', ax=ax)
df.plot(kind='line',x='month_number',y='bathingsoap', color='green', ax=ax)
df.plot(kind='line',x='month_number',y='shampoo', color='yellow', ax=ax)
df.plot(kind='line',x='month_number',y='moisturizer', color='orange', ax=ax)

plt.show()
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
plt.figure(figsize=(10,6))

plt.bar(df['month_number']+0.0 , df['facecream'],color='m',width=0.40,label='facecream')
plt.bar(df['month_number']+0.40, df['facewash'],color='c',width=0.40,label='facewash')
plt.legend()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
ventas_df = df.drop(['month_number', 'total_units', 'total_profit'], axis=1)
sns.boxplot(data=ventas_df)
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
plt.pie(df[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']].sum(), labels=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
```
